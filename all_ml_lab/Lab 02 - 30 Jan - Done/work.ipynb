```python
plt.figure(figsize=(8, 6))
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], alpha=0.7)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'], alpha=0.7)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
sns.pairplot(df, hue='Species')
plt.show()
```

```python
plt.figure(figsize=(12, 10))

# Sepal Length
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='SepalLengthCm', hue='Species', kde=True, bins=20)
plt.title('Sepal Length Distribution')

# Sepal Width
plt.subplot(2, 2, 2)
sns.histplot(data=df, x='SepalWidthCm', hue='Species', kde=True, bins=20)
plt.title('Sepal Width Distribution')

# Petal Length
plt.subplot(2, 2, 3)
sns.histplot(data=df, x='PetalLengthCm', hue='Species', kde=True, bins=20)
plt.title('Petal Length Distribution')

# Petal Width
plt.subplot(2, 2, 4)
sns.histplot(data=df, x='PetalWidthCm', hue='Species', kde=True, bins=20)
plt.title('Petal Width Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))

# Sepal Length
plt.subplot(2, 2, 1)
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Sepal Length by Species')

# Sepal Width
plt.subplot(2, 2, 2)
sns.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Sepal Width by Species')

# Petal Length
plt.subplot(2, 2, 3)
sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Petal Length by Species')

# Petal Width
plt.subplot(2, 2, 4)
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Petal Width by Species')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Sepal Length
plt.subplot(2, 2, 1)
sns.violinplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Sepal Length Distribution by Species')

# Sepal Width
plt.subplot(2, 2, 2)
sns.violinplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Sepal Width Distribution by Species')

# Petal Length
plt.subplot(2, 2, 3)
sns.violinplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Petal Length Distribution by Species')

# Petal Width
plt.subplot(2, 2, 4)
sns.violinplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Petal Width Distribution by Species')

plt.tight_layout()
plt.show()
```
