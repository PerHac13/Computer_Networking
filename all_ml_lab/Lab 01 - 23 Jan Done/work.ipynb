```python
import numpy as np

# 3 arrays of 3 elements each with 0
empty_array = np.empty((3, 3))

full_array = np.full((4, 3), 1) # 4x3 array with all elements as 1
print("Empty Array:\n", empty_array)
print("Full Array:\n", full_array)

```

```python
random_array = np.random.randn(25) # sampled from standard normal distribution
print("Random Array:\n", random_array)
```

```python

array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])
dot_product = np.dot(array1, array2)
print(array1, "\n\n", array2, "\n\n")
print("Dot Product:\n", dot_product)
```

```python

array_to_sort = np.array([[3, 7, 1], [10, 3, 2], [5, 6, 7]])
sorted_row = np.sort(array_to_sort, axis=1) # row-wise ( axis = 1 ) 
sorted_col = np.sort(array_to_sort, axis=0) # col-wise ( axis = 0 )
sorted_whole = np.sort(array_to_sort.flatten())
print("Row-wise Sorted:\n", sorted_row)
print("Column-wise Sorted:\n", sorted_col)
print("Whole Array Sorted:\n", sorted_whole)

```

```python

list_of_arrays = [np.array([1, 2, 3]), np.array([4, 5, 6]), np.array([7, 8, 9])]
mean_list = [np.mean(arr) for arr in list_of_arrays]
print("Mean List:\n", mean_list)

```

```python
total_profit = read_csv_column('C:\\AIML\\ML Lab\\ompany_sales_data.csv', 'total_profit')
months = list(range(1, len(total_profit) + 1))


plt.figure()
plt.plot(months, total_profit)
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Monthly Total Profit")
plt.show()

```

```python

plt.figure()
plt.plot(months, total_profit, linestyle='dotted', color='red', marker='o', markerfacecolor='red', linewidth=3)
plt.xlabel("Month Number")
plt.ylabel("Sold Units Number")
plt.legend(["Profit"], loc='lower right')
plt.title("Monthly Total Profit with Style")
plt.show()
```

```python
toothpaste_sales = read_csv_column('ompany_sales_data.csv', 'toothpaste')
plt.figure()
plt.scatter(months, toothpaste_sales)
plt.grid( linestyle='--')
plt.xlabel("Month Number")
plt.ylabel("Toothpaste Sales")
plt.title("Toothpaste Sales Data")
plt.show()

```

```python
face_cream_sales = read_csv_column('ompany_sales_data.csv', 'facecream')
face_wash_sales = read_csv_column('ompany_sales_data.csv', 'facewash')
x_indexes = np.arange(len(months))
width = 0.3

plt.figure()
plt.bar(x_indexes, face_cream_sales, width=width, label='Face Cream')
plt.bar(x_indexes + width, face_wash_sales, width=width, label='Face Wash')
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.title("Face Cream and Face Wash Sales")
plt.legend()
plt.show()
```
